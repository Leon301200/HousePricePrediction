```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data/preprocessed/train with city.csv')
data
```

```python
plt.hist(data['TARGET(PRICE_IN_LACS)'], bins=10, edgecolor='black')
plt.title("Distribution des prix")
plt.xlabel("Prix")
plt.ylabel("Distribution")
plt.show()
```

```python
plt.scatter(data['SQUARE_FT'], data['TARGET(PRICE_IN_LACS)'])
plt.title("Prix en fonction de la surface")
plt.xlabel("Surface")
plt.ylabel("Prix")
plt.show()
```

```python
sns.boxplot(x='CITY', y='TARGET(PRICE_IN_LACS)', data=data)
plt.title("Prix en fonction des villes")
plt.xlabel("Ville")
plt.ylabel("Prix")
plt.show()
```
