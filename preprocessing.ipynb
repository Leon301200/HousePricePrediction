```python
import pandas as pd
import numpy as np
```

```python
filepath = 'data/train.csv'
```

```python
data = pd.read_csv(filepath)
data
```

```python
# Typage des données 
def can_be_converted_to_bool(col):
    return col.isin([0, 1]).all()

for col in data.columns:
    if can_be_converted_to_bool(data[col]):
        data[col] = data[col].astype(bool)

data
```

```python
# Supprimer les lignes avec plus de deux virgules dans les adresse 
df = data.copy()
df = df[df['ADDRESS'].str.count(',') == 1]
df[['DISTRICT', 'CITY']] = df['ADDRESS'].str.split(',', expand=True)
df = df.drop(['ADDRESS'], axis=1)
df = df.drop(['DISTRICT'], axis=1)
df = df.drop(['CITY'], axis=1)
df
```

```python
df = pd.get_dummies(df, drop_first=True)
df 
```

```python
# Enregistrement du dataSet
df.to_csv('data/preprocessed/train.csv', index=False)
```
